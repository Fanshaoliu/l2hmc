```python
def plot_grid(S, width=8):
    sheet_width = width
    plt.figure(figsize=(8, 8))
    for i in xrange(S.shape[0]):
        plt.subplot(sheet_width, sheet_width, i + 1)
        plt.imshow(S[i], cmap='gray')
        plt.grid('off')
        plt.axis('off')
```
